```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
from matplotlib.cbook import get_sample_data
import matplotlib.colors as mcolors
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy.io.shapereader as shpreader
import shapely.geometry as sgeom
import cartopy.io.img_tiles as cimgt
from matplotlib.offsetbox import AnchoredText
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import warnings
warnings.filterwarnings('ignore')
```

```python
Mbox_lon=np.array([-95,-95,-87,-87,-95])
Mbox_lat=np.array([35,40,40,35,35])

plt.figure(figsize=(10,5))
ax=plt.axes(projection=ccrs.Robinson(central_longitude=-90))
ax.stock_img()
ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)
ax.coastlines(resolution='110m')

plt.plot(Mbox_lon, Mbox_lat, color='blue', linewidth=2, marker=".",
         transform=ccrs.Geodetic(),
)
# Zoom into the contiguous United States
ax.set_extent([-125, -66.5, 20, 50], crs=ccrs.Geodetic())

# Add state boundaries
ax.add_feature(cfeature.STATES, edgecolor='black')

# Zoom into the blue box area with state boundaries
ax.set_extent([Mbox_lon.min(), Mbox_lon.max(), Mbox_lat.min(), Mbox_lat.max()], crs=ccrs.Geodetic())
ax.add_feature(cfeature.STATES, edgecolor='black') # Ensure state boundaries are still added after zooming

plt.show()
```
